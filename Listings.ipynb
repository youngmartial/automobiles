```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select the top 5 neighborhoods
top_5_neighborhoods = unique_hosts_per_neighbourhood.head(5)['neighbourhood']

# Filter the data for the top 5 neighborhoods
top_5_data = df_listings[df_listings['neighbourhood'].isin(top_5_neighborhoods)]

# Calculate the average price for each category of listings within the top 5 neighborhoods
average_price_per_category = top_5_data.groupby(['neighbourhood', 'room_type'])['price'].mean().reset_index()

# Plot the grouped bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x='neighbourhood', y='price', hue='room_type', data=average_price_per_category, palette='muted')
plt.title('Average Price of Listings by Room Type in Top 5 Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Average Price')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Room Type')
plt.tight_layout()
plt.show()
```

```python
df_listings = pd.DataFrame({'number_of_reviews': [50, 100, 200, 300, 400, 750], 'calculated_host_listings_count': [7, 11, 45, 70, 100, 151]})
# create an empty list
levels = []
# loop through the dataframe rows using iterrows()
for index, row in df_listings.iterrows():
  if row['number_of_reviews'] > 300 and row['calculated_host_listings_count'] == 151:
    # add a certain value to the empty list created above
    levels.append('Super Host')
  else:
    # or, add a different value to the empty list created above
    levels.append('Regular Host')
# create new column in the dataframe, and assign the list that was created above, and populated using the for loop
df_listings['Classify_Host'] = levels
print(df_listings)
```
